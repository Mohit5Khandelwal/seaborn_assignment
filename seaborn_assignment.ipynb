Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular data visualization library in Python that provides a high-level interface for creating informative and attractive statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. Scatter Plot:
   - Uses: Scatter plots are useful for visualizing the relationship between two continuous variables. They can reveal patterns, trends, clusters, or correlations in the data.

2. Line Plot:
   - Uses: Line plots are effective for visualizing trends and patterns in data over time. They are commonly used for time series analysis, showing the evolution of a variable over a continuous interval.

3. Bar Plot:
   - Uses: Bar plots are used to compare categorical data and display the distribution of a variable across different categories. They are suitable for visualizing counts, proportions, or averages of different categories.

4. Histogram:
   - Uses: Histograms provide a visual representation of the distribution of a single variable. They group data into bins and show the frequency or count of observations within each bin, allowing you to understand the shape and spread of the data.

5. Heatmap:
   - Uses: Heatmaps are effective for visualizing matrix-like data, where each cell's color represents the magnitude of the variable. They are useful for correlation matrices, confusion matrices, or any data that can be represented as a matrix.


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = 
"timepoint" and y = "signal" for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset( 'fmri' )
```

```python
df.head()
```

```python
sns.lineplot( x = 'timepoint' , y = 'signal' , data = df , hue='event' , style = 'region')
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 
'pclass', y = 'age' and y = 'fare'.

```python
# Loading the titanic data set 
df = sns.load_dataset( 'titanic' )
```

```python
# Box Plot using seaborn library 
sns.boxplot( x = 'pclass' , y = 'age' , data = df , width = 0.4 )
```

```python
sns.boxplot( x = 'pclass' , y = 'fare' , data = df  )
```
